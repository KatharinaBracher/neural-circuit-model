## Explorig Paramater K, initI, $\sigma$, tau

```python
#K     higher K - steeper slope, std more different (smaller for small stim, larger for large stim)
#initI Slope, location, std (bigger difference when higher)
#sigma std, slope if very high
#tau   std, slope (the lower the steeper) ~ which stim is on the identiy line (the lower tau, the lower stim on id)
```

```python
fig, ax = plt.subplots(4,5, figsize=(16,10), sharex=True, sharey=True)
stimuli = [400, 450, 500, 550, 600, 650, 700]

K_lst =[3,4,5,6]
tau=[80,90,100,110,120]

i = 0
for K in K_lst:
    for t in tau:
                       
        initI = 0.76                     
        PARAMS_DICT['sigma'] = 0.01       
        PARAMS_DICT['tau'] = t 
        task = otg.OneTwoGo_Task(PARAMS_DICT)
        task.plot_behavior(stimuli, K, initI, ax.flatten()[i])
        ax.flatten()[i].text(np.min(stimuli)-100, np.max(stimuli)+100, 'K='+str(K)+', tau='+str(t))
        i += 1
        
for a in ax.flat:
    a.set(xlabel='Stimulus (ms)', ylabel='Production (ms)')
    a.label_outer()

        
fig.text(0.5, 1.0, 'tau', ha='center', fontsize=24)
fig.text(-0.03, 0.5, 'K', va='center', rotation='vertical',  fontsize=24)

plt.tight_layout()
```
