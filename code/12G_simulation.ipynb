```python
stimuli = [400, 500, 600, 700, 800, 900, 1000]

K = 5.0                           #Slope
initI = 0.77                      #Slope, hight, std
PARAMS_DICT['sigma'] = 0.009      #std

PARAMS_DICT['tau'] = 120  #which stim is on the identiy line (the lower tau, the lower stim are in id)
task = otg.OneTwoGo_Task(PARAMS_DICT)
task.plot_behavior(stimuli, K, initI)
```
