```python
def mean_I(sorted_data):
    mean_I = []
    for s in range(7):
        summe = 0
        for i in range(len(sorted_data.I_sorted[s])):
            summe += sorted_data.I_sorted[s][i][1][0]
        mean_I.append(summe/len(sorted_data.I_sorted[s]))
    return mean_I
```
